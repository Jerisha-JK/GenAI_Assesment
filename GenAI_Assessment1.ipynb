```python
from groq import Groq
import json

# Initialize client
client = Groq(api_key="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")

# Messages for a single question
messages = [
    {"role": "system", "content": "You are a converter who generates content in JSON or bullet points."},
    {"role": "user", "content": "Provide employee details"}
]

# Create completion
completion = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=messages,
    temperature=0.3,
    max_completion_tokens=1024,
    top_p=1.0,
    stream=True
)

# Collect streaming output
response_text = ""
for chunk in completion:
    content = chunk.choices[0].delta.content or ""
    print(content, end="")  # optional: print live
    response_text += content
print("\n")  # end of streaming output

# Try parsing JSON
try:
    response_json = json.loads(response_text)
except json.JSONDecodeError:
    response_json = {"error": "Response is not valid JSON", "raw_response": response_text}

# Print formatted JSON
print("Parsed JSON response:")
print(json.dumps(response_json, indent=8))
```

```python
import os
from groq import Groq

# Load template from file
def load_template(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        return f.read()

# Fill placeholders in the template
def fill_prompt(template, topic, tone):
    return template.format(topic=topic, tone=tone)

# Send prompt to Groq
def send_to_groq(prompt):
    client = Groq(api_key="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")

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message["content"]

# Main workflow
if __name__ == "__main__":
    template = load_template("prompt_template.txt")

    # Dynamic values
    topic = "Latest AI advancements"
    tone = "professional"

    final_prompt = fill_prompt(template, topic, tone)
    print("Final Prompt:\n", final_prompt)

    output = send_to_groq(final_prompt)
    print("\nGroq Response:\n", output)
```
